```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

import pandas as pd
```

```python
def driver_setup():
    #setup the driver
    driver = webdriver.Chrome()
    return driver
```

```python
def scrape_twitter_data(driver, url, num_tweets=50):
    # Open Twitter
    driver.get(url)

    time.sleep(3)  # Allow the page to load

    # Find and fill in the username and password fields
    username_field = driver.find_element(By.XPATH , "/html/body/div/div/div/div[1]/div/div/div/div/div/div/div[2]/div[2]/div/div/div[2]/div[2]/div/div/div/div[5]/label/div/div[2]").click()
    time.sleep(1)
    click_field = driver.find_element(By.XPATH, "/html/body/div/div/div/div[1]/div/div/div/div/div/div/div[2]/div[2]/div/div/div[2]/div[2]/div/div/div/div[5]/label")
    click_field.send_keys("2019205659")
    
    driver.find_element(By.XPATH, "/html/body/div/div/div/div[1]/div/div/div/div/div/div/div[2]/div[2]/div/div/div[2]/div[2]/div/div/div/div[6]/div/span/span").click()
    
    time.sleep(2)
    password_field = driver.find_element(By.XPATH, "/html/body/div/div/div/div[1]/div/div/div/div/div/div/div[2]/div[2]/div/div/div[2]/div[2]/div[1]/div/div/div[2]/div/label/div/div[2]").click()
    Pclick_field = driver.find_element(By.XPATH, "/html/body/div/div/div/div[1]/div/div/div/div/div/div/div[2]/div[2]/div/div/div[2]/div[2]/div[1]/div/div/div[2]/div/label")
    #Password comes in the blank  space
    Pclick_field.send_keys("  ")
    
    #log in
    login_button = driver.find_element(By.XPATH, '//span[text()="Log in"]')
    login_button.click()

    # Scroll down to load more tweets
    for _ in range(num_tweets // 10 ):  # Scrolling 5 times 
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(15)

    # Find tweet descriptions and keywords
    tweet_elements = driver.find_elements(By.XPATH, "//div[contains(@class, 'css-1dbjc4n') and contains(@class, 'r-1iusvr4') and contains(@class, 'r-16y2uox') and contains(@class, 'r-1777fci') and contains(@class, 'r-kzbkwu')]")

    tweets_data = []
    IDnum = 1
    for tweet_element in tweet_elements[:num_tweets]:
        # Description
        description_element = tweet_element.find_element(By.CSS_SELECTOR, '[data-testid="tweetText"]')
        description = description_element.text.strip()

        # Keyword
        
        keyword = ""
        try:
            keyword_element = tweet_element.find_element(By.CSS_SELECTOR, 'span.css-901oao.css-16my406.r-poiln3.r-b88u0q.r-bcqeeo.r-qvutc0')
            time.sleep(1)
            keyword = keyword_element.text.strip()
        
        except NoSuchElementException:
            pass
        

        tweets_data.append({"ID": IDnum, "Keyword": keyword, "Description": description})
        
        IDnum += 1
        


    driver.quit()
    #putting data in panda dataframe anf saving data in CSV 
    csv_filename = "test.csv"
    df = pd.DataFrame(tweets_data)
    df.to_csv(csv_filename, index=False)
    return df
```
